```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
sanjuan_dataset = pd.read_csv("../../dengueData/SanJuan/san_juan_training_data.csv")
print(sanjuan_dataset.head())
print("Number of observations: ", len(sanjuan_dataset))
```

```python
sanjuan_cases = sanjuan_dataset["total_cases"].values
print(sanjuan_cases.shape)
```

```python
from sklearn.metrics import mean_squared_error

lag_weeks = 0
window_size = 4
split_index = int(len(sanjuan_cases)* 0.7)

test = sanjuan_cases[split_index+lag_weeks+window_size:]
predictions = sanjuan_cases[split_index+window_size-1:-lag_weeks-1] # 1 week behind

print(test.shape)

print(sanjuan_cases[-10:])
print(test[-10:])
print(predictions[-10:])
print(predictions.shape)
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
```
